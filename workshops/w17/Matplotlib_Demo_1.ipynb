# Clustering Algorithms Visualization


```python
from sklearn.datasets import make_blobs

# Generate synthetic data 
X, y = make_blobs(n_samples=300, centers=4, random_state=45)
```

## Hierarchical Clustering

Hierarchical clustering builds a tree of clusters.

In this code, we perform hierarchical clustering on synthetic data (X) using the AgglomerativeClustering algorithm with four clusters. After obtaining cluster labels (h_labels), we visualize the results by creating a scatter plot using Matplotlib. Each data point is colored according to its assigned cluster, providing insight into the hierarchical clustering structure.

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# Apply Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=4)
h_labels = hierarchical.fit_predict(X)
```

```python
# Visualize Hierarchical Clustering
plt.scatter(X[:, 0], X[:, 1], c=h_labels, cmap='viridis', edgecolors='k', s=50)
plt.title('Hierarchical Clustering')
plt.show()
```

## K-means Clustering 

K-Means clustering is applied to synthetic data (X) using the KMeans algorithm with four clusters. After obtaining cluster labels (k_labels), the results are visualized through a scatter plot using Matplotlib. Each data point is colored based on its assigned cluster, and red asterisk markers represent the centroids of each cluster. 

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
k_labels = kmeans.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=k_labels, cmap='viridis', edgecolors='k', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='o', s=200, label='Centroids')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```

## Spectral Clustering

After obtaining cluster labels (s_labels), the results are visualized through a scatter plot using Matplotlib. Each data point is colored based on its assigned cluster, providing an understanding of the clustering structure identified by the spectral clustering algorithm with four clusters.

```python
from sklearn.cluster import SpectralClustering
```

```python
spectral = SpectralClustering(n_clusters=4, affinity='nearest_neighbors', random_state=42)
s_labels = spectral.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=s_labels, cmap='viridis', edgecolors='k', s=50)
plt.title('Spectral Clustering')
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
```

```python
fig, ax = plt.subplots()
x_data = np.arange(0, 10, 0.1)
line, = ax.plot(x_data, np.sin(x_data))

def update(frame):
    line.set_ydata(np.sin(x_data + frame / 10))
    return line,

ani = FuncAnimation(fig, update, frames=100, interval=50)
```

## Part 2: Loading and Manipulating Images

```python
%matplotlib inline
import warnings; warnings.filterwarnings('ignore')
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt

img = load_sample_image("flower.jpg")
plt.imshow(img)
```

Graphing color distributions of an imgae.

```python
# import seaborn as sns
fig = plt.figure(figsize = (16, 4))
plt.subplot(131)
_ = plt.hist(img[:, :, 0].flatten(), bins = 25, range = (0, 255), color = 'r')
plt.subplot(132)
_ = plt.hist(img[:, :, 1].flatten(), bins = 25, range = (0, 255), color = 'g')
plt.subplot(133)
_ = plt.hist(img[:, :, 2].flatten(), bins = 25, range = (0, 255), color = 'b')
```

```python
fig = plt.figure(figsize = (16, 4))
plt.subplot(131)
plt.title("Red")
plt.imshow(img[:, :, 0], cmap = "gray")
plt.subplot(132)
plt.title("Green")
plt.imshow(img[:, :, 1], cmap = "gray")
plt.subplot(133)
plt.title("Blue")
plt.imshow(img[:, :, 2], cmap = "gray")
```

Thresholding with Matplotlib and Numpy.

```python
green_channel = img[:, :, 1]  # pull out the green channel, just so we don't have to keep indexing it
binarized = (green_channel > np.median(green_channel))
plt.imshow(binarized, cmap = "gray")
```

Adding guassian blur effects to images.

```python
from scipy import ndimage
# Note the two different "sigma" values in the next 2 lines.
blurred = ndimage.gaussian_filter(green_channel, sigma = 3)
very_blurred = ndimage.gaussian_filter(green_channel, sigma = 6)

fig = plt.figure(figsize = (16, 4))
plt.subplot(131)
plt.title("Original")
plt.imshow(green_channel, cmap = "gray")
plt.subplot(132)
plt.title("Sigma = 3")
plt.imshow(blurred, cmap = "gray")
plt.subplot(133)
plt.title("Sigma = 6")
plt.imshow(very_blurred, cmap = "gray")
```

Image segmentation using ndimage.label().

The function **`ndimage.label()`** takes a binary image as input, and finds all the *connected components* in the image.

Connected components are contiguous blocks of pixels with the value "1". The function finds these blocks and assigns each one a unique ID, thereby picking out all the "objects" in the image.

```python
binarized = (blurred > blurred.mean())
labels, num_labels = ndimage.label(binarized)
print("Found {} objects.".format(num_labels))

fig = plt.figure(figsize = (14, 8))
plt.subplot(121)
plt.imshow(binarized, cmap = "gray")
plt.subplot(122)
plt.imshow(labels)
```

```python
# This gives us a list of region sizes.
obj_sizes = ndimage.sum(binarized, labels, range(num_labels + 1))

# Now we define a threshold on the SIZE of the object.
remove_pixels = (obj_sizes < obj_sizes.max())

# Create a mask on the size, and set all the objects that fit this mask to 0.
mask_labels = remove_pixels[labels] # Fancy indexing!
labels[mask_labels] = 0 # Boolean indexing!
plt.imshow(labels, cmap = "gray")
```

```python
# Our original RGB image is in the variable "img"
img_nobackground = np.copy(img)
img_nobackground[labels == 0, :] = 0
# "labels" is 2D, but "img" is 3D, so that third dimension is covered with ":"
plt.imshow(img_nobackground)
```
